```python
import http.client
import json
import time

# consume coincap /rates api
conn = http.client.HTTPSConnection("api.coincap.io")
payload = ''
headers = {'Content-type': 'application/json'}
conn.request("GET", "/v2/rates/bitcoin", payload, headers)
res = conn.getresponse()
data = res.read()

# write to file
file_name = 'coincap_' + time.strftime("%Y%m%d-%H%M%S") + '.json'
file_write = open(file_name, 'w')
file_write.write(json.dumps(data.decode('utf-8')))
file_write.close()

json_object = json.loads(data.decode("utf-8"))
print("response from coincap:")
print(json.dumps(json_object, indent=2))
```
